```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with pd.ExcelFile(r"D:\小鸡理财\百度云同步盘\小鸡理财\数据报告\用户类报告\流失用户分析\流失用户属性分析.xlsx") as xlsx:
    df = pd.read_excel(xlsx,'表格')
    df_tz = pd.read_excel(xlsx,'投资')
    df_id = pd.read_excel(xlsx,'身份认证')
```

```python
# df_new_member = pd.read_csv(r'D:\小鸡理财\百度云同步盘\小鸡理财\数据报告\用户类报告\流失用户分析\运营数据.txt',
#                             delimiter="\t",
#                             header=None,
#                             engine = 'python',
#                             names=['用户名','手机号码','身份证'])
```

身份证标准化

```python
df_id['id'] = df_id['身份证'].str.upper()
```

'投资' 与 '身份认证' 合并

```python
df_tz_new = pd.merge(df_tz,df_id,
                    left_on = '投资用户',right_on = '用户名',
                    how = 'left')
```

```python
df_tz_new.info()
```

去掉空值

```python
df_tz_new.dropna(subset=['id'],inplace=True)
df_tz_new.shape
```

```python
df_tz_new.drop(columns=['1标期限','月or天','用户名','身份证'], inplace=True)
```

```python
df_tz_new.head()
```

```python
df_tz_new = pd.merge(df_tz,df_id,
                    left_on = '投资用户',right_on = '用户名',
                    how = 'left')
```

记录每个id的最后一次投资时间

```python
df_tz_new['投资时间'] = pd.to_datetime(df_tz_new['投资时间'])
```

```python
df_tz_new.sort_values(by=['投资时间','id'],ascending=False,inplace=True)
```

```python
df_last_invest = df_tz_new.drop_duplicates(subset=['id'])
```

```python
df_last_invest
```

合并 最后投资时间 & 注册来源 到表格df

```python
df_1 = pd.merge(df,df_last_invest,
               on='id',
               how='left')
```

```python
df_1.head()
```

```python
df_1.columns
```

```python
df_1 = df_1.loc[:,['用户名_待收', '真实姓名_x', '发生时间', '注册时间_x', 'id','预计流失时间', '流失or激活','标类型', '期限','投资时间_y', 'new_邀请渠道']]
```

```python
df_1.rename(columns={'投资时间_y' : '最后投资时间'},inplace=True)
```

计算每个用户的累计年化投资额 & 累计投资次数

```python
gp_user_tz = df_tz_new.groupby('id').agg({'年化投资额': np.sum, '投资用户': np.size}).reset_index()
```

```python
gp_user_tz.head()
```

合并 累计年化投资额 & 投资次数 到表格df_1

```python
df_2 = pd.merge(df_1,gp_user_tz,
               on='id',
               how='left')
```
