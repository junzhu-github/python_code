# 饼图-玫瑰图

```python
from pyecharts import Pie

attr = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
v1 = [19,21,32,20,20,33]
v2 = [19,21,32,20,20,33]

pie = Pie('饼图-玫瑰图示例',title_pos = 'center',width = 900)
pie.add('商品A',
        attr,
        v1,
        center = [25,50],
        is_random = True,
        radius = [30,75],
        rosetype='radius')

pie.add('商品B',
        attr,
        v1,
        center = [75,50],
        is_random = True,
        radius = [30,75],
        rosetype='area',
       is_legend_show=False,
       is_label_show=True)

pie
```

# 条形-折现图

```python
from pyecharts import Bar,Line,Overlap

attr = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
v1 = [19,21,32,20,20,33]
v2 = [29,31,42,30,30,43]
v3 = [9,11,32,30,30,43]

bar = Bar('Line - Bar 示例')
bar.add('bar',attr,v1)

line = Line()
line.add('line',attr,v2)

overlap = Overlap()
overlap.add(bar)
overlap.add(line)

overlap
```
