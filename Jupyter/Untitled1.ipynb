```python
df = pd.read_csv('02.csv')

df.head()
```

```python
gp = df.groupby('手机号')['手机号'].count()

gp_1 = gp[gp < 2 ]

gp_1.head()

l = list(gp_1.index)

print(type(l))
```

```python
df_new_1 = df.loc[df['标类型'].str.contains('新手') ,['标类型','期限','总投资额','手机号','注册时间','new_邀请渠道']]

df_new_1.head()
```

```python
df_new_2 = df.loc[df['期限'] == '7天',['标类型','期限','总投资额','手机号','注册时间','new_邀请渠道']]

df_new_2.head()
```

```python
df_new = pd.concat([df_new_1 , df_new_2]).drop_duplicates()
    
df_new.head()
```

```python
df_final = df_new[df['手机号'].isin(l)]

df_final.head()
```
