```python
df1 = pd.read_csv('定存标投资记录.CSV',engine = 'python') 
df2 = pd.read_csv('月月升标投资记录.CSV',engine = 'python') 
df3 = pd.read_csv('月月升标续投记录.CSV',engine = 'python') 
df4 = pd.read_csv('债权转让标投资记录.CSV',engine = 'python') 
df5 = pd.read_csv('存管标投资记录.CSV',engine = 'python') 
```

```python
df_list =[df1,df2,df3,df4,df5] 
```

```python
list(map(lambda x: print(x.columns),df_list))
```

```python
df = pd.DataFrame()
df_list_not_df3 = [df1,df2,df4,df5]


for i in df_list_not_df3:
    df = df.append(i[['总投资额', '投资时间']])

df3_temp = df3[['总投资额', '续投时间']]
df3_temp.columns = ['总投资额', '投资时间']
# df3_temp.head()
df = df.append(df3_temp)

df['mark'] = [i for i in range(len(df))]
df = df.set_index(df['mark'])

df.info()

df.head()
```

```python
temp = df['投资时间'].str.split(' ',expand = True)
temp.columns = ['日期','时间']

temp['mark'] = [i for i in range(len(temp))]
temp = temp.set_index(df['mark'])

temp.info()
temp.head()
```

```python
df_invest = df.merge(temp)
df_invest.drop('mark',axis = 1,inplace = True)

df_invest.info()

df_invest.head()
```

```python
df_invest['日期'] = pd.to_datetime(df_invest['日期'])

df_invest.info()
```

```python
gp_invest = df_invest.groupby('日期')['总投资额'].sum()

gp_invest
```

```python
gp_sum = gp_invest.resample('M', level='日期').sum()

gp_sum
```

```python
label_list = ['1月', '2月', '3月', '4月', '5月', '6月']    # 横坐标刻度显示值
num_list1 = gp_sum.values / 10000000      # 纵坐标值1
x = range(len(num_list1))

"""
绘制条形图
left:长条形中点横坐标
height:长条形高度
width:长条形宽度，默认值0.8
label:为后面设置legend准备
"""
rects1 = plt.bar(left=x, height=num_list1, width=0.8, alpha=0.8, color='red', label="投资额")
plt.ylim(0, 15)     # y轴取值范围
plt.ylabel("单位：千万")

"""
设置x轴刻度显示值
参数一：中点坐标
参数二：显示值
"""
plt.xticks([index + 0.2 for index in x], label_list)
plt.xlabel("月份")
plt.title('小鸡')
plt.legend()     # 设置题注

# 编辑文本
for rect in rects1:
    height = rect.get_height()
#     print('=',height,'=')
    plt.text(rect.get_x() + rect.get_width() / 2, height+1, str(round(height*1000)) , ha="center", va="bottom")

plt.show()
```
