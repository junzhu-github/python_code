```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel(r'D:\小鸡理财\百度云同步盘\小鸡理财\数据报告\月报\历史数据.xlsx','投资')
```

```python
df.head()
```

```python
df['手机号'].isnull().value_counts()
```

```python
df['手机号'].dropna(inplace=True)
```

```python
df.info()
```

```python
df['time'] = pd.to_datetime(df['投资时间']).dt.strftime('%Y-%m-%d')
```

```python
df['time'].head()
```

```python
mask = (df['time'] >= '2018-01-01') & (df['time'] <= '2018-08-31')
```

```python
mask.tail()
```

```python
df_temp = df.loc[mask,['手机号','time']]
```

```python
df_temp[df_temp.time == '2018-08-31']
```

```python
df_temp.sort_values(by = ['time'],inplace=True,ascending=True)
```

```python
df_temp.shape
```

```python
df_temp.drop_duplicates(subset=['手机号','time'], keep='first', inplace=True)
```

```python
df_temp.shape
```

```python
df_temp['time'] = pd.to_datetime(df_temp['time'])
```

```python
df_temp.info()
```

```python
gp = df_temp.groupby('手机号',as_index=False)['time'].transform(lambda x: x.shift(-1) - x)
```

```python
gp['手机号'] = df_temp['手机号']
```

```python
gp['手机号'] = gp['手机号'].astype('str')
```

```python
gp.shape
```

```python
gp[gp['手机号']=='13001728856.0']
```

```python
gp.dropna(inplace=True)
```

```python
gp['time'] = gp['time'].astype('timedelta64[D]')
```

```python
gp['time'].hist(bins=50)
```

```python
tp = gp.groupby('time').count()
```

```python
tp
```

```python
tp.to_excel('用户投资时间间隔.xlsx')
```
